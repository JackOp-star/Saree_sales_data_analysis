```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("uncleaned_saree_sales_dataset.csv")
```

```python
df
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
```

```python
df
```

```python
df['Region'].unique()
```

```python
df['Region'] = df['Region'].str.title()
```

```python
df['Region'].unique()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors= 'coerce')
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
new_median = ["Quantity","Unit_Price","Sales"]

for new in new_median:
    df[new] = df[new].fillna(df[new].median())
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["Saree_Type","Color"])
```

```python
df.isnull().sum()
```

```python
df
```

```python
df = df[df['Quantity']>0]
```

```python
df
```

```python
df = df[df['Unit_Price']>0]
```

```python
df
```

```python
neg = df[df['Unit_Price']<=0]
```

```python
neg
```

```python
df["saless"] = df['Quantity'] * df['Unit_Price']
```

```python
df
```

```python
df['Final_Sales'] = df['saless'] - (df['saless'] * df['Discount_(%)']/100)
```

```python
df
```

```python
df =df.drop('Sales',axis=1)
```

```python
Q1 = df["Final_Sales"].quantile(0.25)
Q3 = df["Final_Sales"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df["Final_Sales"] >= lower) & (df["Final_Sales"] <= upper)]
```

```python
df
```

```python
df['Year'] =df['Order_Date'].dt.year.fillna(0).astype(int)
```

```python
df['Month'] = df['Order_Date'].dt.month.fillna(0).astype(int)
```

```python
df
```

```python
df.dtypes
```

```python
df.to_csv("new_cleaned_saree_sales.csv",index=False)
```

```python
import matplotlib.pyplot as plt
```

```python
monthly_sales = df.groupby('Month')['Final_Sales'].sum()

plt.figure(figsize=(12,5))
plt.plot(monthly_sales.index,monthly_sales.values,marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.grid()
plt.show()
```

```python
yearly_sales = df.groupby('Year')['Final_Sales'].sum()


plt.bar(yearly_sales.index, yearly_sales.values)
plt.title("Yearly Revenue Growth")
plt.xlabel("Year")
plt.ylabel("Total Revenue")
plt.show()
```

```python
df
```

```python
festive = df.groupby('Occasion')['Final_Sales'].sum()
plt.bar(festive.index, festive.values)
plt.xticks(rotation=45)
plt.title("Revenue by Occasion")
plt.show()
```

```python
sare_types = df.groupby('Saree_Type')['Final_Sales'].sum()

plt.bar(sare_types.index,sare_types.values)
plt.xticks(rotation=40)
plt.title("Top Selling Saree Types")
plt.show()
```

```python
profit = 0.25
df['Profit'] = df['Final_Sales'] * profit
```

```python
df
```

```python
monthly_profit = df.groupby('Month')['Profit'].sum()

plt.figure(figsize=(8,4))
plt.plot(monthly_profit.index,monthly_profit.values,marker='o',mfc='red',color ='hotPink')
plt.title('Monthly Profit Trend')
plt.xlabel('Month')
plt.ylabel('Total_Profit')
plt.grid()
plt.show()
```
